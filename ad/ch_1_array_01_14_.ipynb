알고리즘 느는 방법은 코드를 그대로 시뮬레이션 해보는 것

#### 버블정렬은 제일큰게 맨 뒤로 가게됨 => sorted!

#### 이런 똑같은 것을 또 반복 또 반복













```python
# 로직이 어떻게 전개 되는지 체득해야 한다.
def BubbleSort(a):
    for i in range(len(a)-1, 0, -1):
        for j in range(0,i):
            if a[j] > a[j+1]: # 바꾼다. 역전. 역전 카운팅으로 뭔가 문제풀 수도 있을듯
                a[j],a[j+1] = a[j+1], a[j]
             
a = [5,3,2,4,1]
BubbleSort(a)
print(a)
```

## 카운팅 정렬
#### 항목들의 순서를 결정하기 위해 집합에 각 항목이 몇 개씩 있는지 세는 작업을 하여 선형 시간에 정렬하는 효율적인 알고리즘
#### 심플하며 적재적소에 잘 쓸 수 있다.

    * 정렬하려는 데이터가 카운팅이 되어야 한다(엘리먼트들이 정수여야 한다.)
    * 범위가 있어야 한다.(배열의 크기를 정하기 위해)

```python
a = [0,4,1,3,1,2,4,1] # 1 ~ 4의 배열을 가짐
b =  [None for __ in range(len(a))]
c = [0 for __ in range(5)] # count 배열 생성(a의 범위를 가진)

for i in range(0, len(a)):
    c[a[i]] += 1
    
for i in range(1, len(c)):
     c[i] += c[i-1]
    
for i in range(len(b)-1,-1,-1):
    b[c[a[i]]-1] = a[i]
    c[a[i]] += - 1
    
# 출현한 순서대로 정렬 / structure 구조일 경우
print(b)
```
